```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
data=pd.read_csv('/content/sample_data/covid_toy.csv')
```

```python
data.head()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(data.drop(columns=['has_covid']),data['has_covid'],test_size=0.2)
```

```python
X_train.head()
```

```python
X_train_tnf=pd.DataFrame(transformer.fit_transform(X_train),columns=('fever','cough','gender','city_1','city_2','city_3','age'))
X_train_tnf.head()
```

```python
X_test_tnf=pd.DataFrame(transformer.fit_transform(X_test),columns=('fever','cough','gender','city_1','city_2','city_3','age'))
X_test_tnf.head()
```
